**üìä An√°lise de Dados *Sobrenatural: A Era Kripke (Temporadas 1-5)***

Neste projeto, exploro a biblioteca Pandas para analisar a evolu√ß√£o da s√©rie Supernatural. O objetivo √© demonstrar compet√™ncias fundamentais de Ci√™ncia de Dados, desde a manipula√ß√£o de estruturas at√© √† visualiza√ß√£o de m√©tricas complexas, utilizando como base o universo dos irm√£os Sam e Dean Winchester.

Neste "Dossier de Ca√ßa Digital", vou focar-me em:

* *Pandas Introduction:* Importa√ß√£o e prepara√ß√£o do ambiente.

* *Pandas Installation:* Configura√ß√£o dos pacotes necess√°rios.

* *DataFrames:* Estrutura√ß√£o de dados sobre epis√≥dios ic√≥nicos.

**An√°lise Tem√°tica:** Estudo de vari√°veis como o estado do Impala, tipos de criaturas enfrentadas e o impacto emocional das temporadas atrav√©s da an√°lise de notas e mortes de personagens.

Para come√ßar a ca√ßada aos dados, precisamos de "chamar" a biblioteca para o nosso c√≥digo.

```python

import pandas as pd

# Criar os dados com temporadas distintas (1 a 5)
dados_supernatural = {
    'Temporada': [1, 2, 3, 4, 5],
    'Epis√≥dio': ['Pilot', 'All Hell Breaks Loose', 'No Rest for the Wicked', 'Lazarus Rising', 'Swan Song'],
    'Vil√£o Principal': ['Woman in White', 'Azazel', 'Lilith', 'Alastair', 'Lucifer'],
    'Nota_IMDb': [8.6, 9.2, 8.9, 9.5, 9.7],
    'Arma_Utilizada': ['Sal/Ferro', 'Colt', 'Faca da Ruby', 'L√¢mina de Anjo', 'Sacrif√≠cio']
}

print("Dados de Supernatural carregados com sucesso!")
```

**Explica√ß√£o:**

Aqui usei *import pandas as pd*. O as pd √© um padr√£o da comunidade que serve para escrevermos menos c√≥digo. Em vez de escrever pandas.DataFrame(), escrevi apenas pd.DataFrame(). Criei tamb√©m um dicion√°rio Python que servir√° de base para a tabela.

**Dataframes**

Vou transformar isto na tabela oficial dos Winchesters e fazer uma an√°lise.

```python
# Criar o DataFrame
df_hunt = pd.DataFrame(dados_supernatural)

# Exemplo Fixe 1: Ordenar por nota (Do melhor para o "pior")
ranking_episodios = df_hunt.sort_values(by='Nota_IMDb', ascending=False)
print("--- Ranking de Epis√≥dios (Melhores Notas) ---")
display(ranking_episodios)

# Exemplo Fixe 2: Contar que armas s√£o mais usadas (fict√≠cio neste set)
uso_armas = df_hunt['Arma_Utilizada'].value_counts()
print("\n--- Invent√°rio de Armas Utilizadas ---")
print(uso_armas)
```

**Explica√ß√£o:**

O comando sort_values organiza a tabela. Usei ascending=False para que o epis√≥dio com a maior nota (9.7 - Swan Song) apare√ßa no topo. O value_counts() permite saber, por exemplo, quantas vezes os rapazes usaram o "Colt" vs "Sal" nas linhas de dados.

**Outros dados**

1.   ***"Morte de Personagem Importante": (Sim/N√£o)*** - Para analisar em que temporadas o impacto emocional foi maior.
2.   ***"Presen√ßa do Impala 67": (Sim/N√£o)*** - Contar as vezes que ele foi destru√≠do!
3. ***"Localiza√ß√£o (Estado dos EUA)":*** O Pandas consegue agrupar dados por regi√£o. Mostrar que a maioria das ca√ßadas acontece no "Midwest".
4. ***"Tipo de Criatura":*** Separar entre Esp√≠ritos, Dem√≥nios, Anjos e Monstros Cl√°ssicos.





```python
import pandas as pd

# 1. Cria√ß√£o do DataFrame com os epis√≥dios do c√≥digo inicial e os novos dados
dados_winchester = {
    'Temporada': [1, 2, 3, 4, 5],
    'Epis√≥dio': ['Pilot', 'All Hell Breaks Loose', 'No Rest for the Wicked', 'Lazarus Rising', 'Swan Song'],
    'Morte_Personagem_Imp': ['N√£o', 'Sim', 'Sim', 'N√£o', 'Sim'], # Mary/Pai, Dean, Castiel/Bobby...
    'Estado_Impala': ['Bom', 'Bom', 'Amolgado', 'Bom', 'Destru√≠do'],
    'Localiza√ß√£o_Estado': ['Kansas', 'Dakota do Sul', 'Indiana', 'Illinois', 'Kansas'],
    'Tipo_Criatura': ['Fantasma', 'Dem√≥nio', 'Hellhound', 'Anjo', 'L√∫cifer'],
    'Nota_IMDb': [8.6, 9.2, 8.9, 9.5, 9.7]
}

df_spn = pd.DataFrame(dados_winchester)

# --- AN√ÅLISE DOS DADOS ---

print("### 1. IMPACTO EMOCIONAL (Mortes Importantes)")
print(df_spn['Morte_Personagem_Imp'].value_counts())

print("\n### 2. INTEGRIDADE DO IMPALA")
# Verifica√ß√£o como o carro sobreviveu a estas temporadas
print(df_spn[['Epis√≥dio', 'Estado_Impala']])

print("\n### 3. MAPA DE CA√áA (Estados)")
# Ver onde a a√ß√£o se concentra
print(df_spn['Localiza√ß√£o_Estado'].unique())

print("\n### 4. BESTI√ÅRIO (Evolu√ß√£o das Amea√ßas)")
# Mostra como as criaturas ficaram mais perigosas com o tempo
display(df_spn[['Temporada', 'Tipo_Criatura']])

# Exemplo: O epis√≥dio com a maior nota onde o Impala foi destru√≠do
print("\n--- O Cl√≠max da S√©rie ---")
climax = df_spn[df_spn['Nota_IMDb'] == df_spn['Nota_IMDb'].max()]
display(climax)
```

**Explica√ß√£o Final:**

*Consist√™ncia:* Manter estrutura original de 5 temporadas e 5 epis√≥dios ic√≥nicos, mas agora cada um carrega informa√ß√µes sobre o Impala, a Localiza√ß√£o e o Tipo de Inimigo.

*A l√≥gica do max():* No √∫ltimo bloco de c√≥digo, usei df_spn['Nota_IMDb'].max(). Ele procura automaticamente o valor mais alto na coluna de notas e mostra que o epis√≥dio "Swan Song" √© o auge dos dados.

*Visualiza√ß√£o Seletiva:* Usei df_spn[['Coluna A', 'Coluna B']] para mostrar apenas o que interessa ao utilizador, sem poluir o ecr√£ com a tabela inteira.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Prepara√ß√£o dos dados
dados_completos = {
    'Temporada': [1, 2, 3, 4, 5],
    'Epis√≥dio': ['Pilot', 'All Hell Breaks Loose', 'No Rest for the Wicked', 'Lazarus Rising', 'Swan Song'],
    'Morte_Personagem_Imp': ['N√£o', 'Sim', 'Sim', 'N√£o', 'Sim'],
    'Estado_Impala': ['Bom', 'Bom', 'Amolgado', 'Bom', 'Destru√≠do'],
    'Localizacao_Estado': ['Kansas', 'South Dakota', 'Indiana', 'Illinois', 'Kansas'],
    'Tipo_Criatura': ['Esp√≠rito', 'Dem√≥nio', 'Monstro Cl√°ssico', 'Anjo', 'Anjo'],
    'Nota_IMDb': [8.6, 9.2, 8.9, 9.5, 9.7]
}

df_spn = pd.DataFrame(dados_completos)

# Configurar o estilo visual
sns.set_theme(style="darkgrid")
plt.figure(figsize=(16, 10))

# --- GR√ÅFICO 1: Impacto Emocional ---
plt.subplot(2, 2, 1)
sns.countplot(data=df_spn, x='Morte_Personagem_Imp', palette=['#8b0000','#0E402D'])
plt.title('1. Impacto Emocional: Mortes Importantes', fontsize=14)
plt.xlabel('Houve Morte de Personagem?')
plt.ylabel('Quantidade de Epis√≥dios')

# --- GR√ÅFICO 2: Presen√ßa e Estado do Impala ---
plt.subplot(2, 2, 2)
# Cria√ß√£o de l√≥gica para contar as vezes que foi destru√≠do
df_spn['Destruido'] = df_spn['Estado_Impala'].apply(lambda x: 'Sim' if x == 'Destru√≠do' else 'N√£o')
sns.countplot(data=df_spn, x='Destruido', palette=['#7C7C7C','#000000' ])
plt.title('2. O Impala foi Destru√≠do?', fontsize=14)
plt.xlabel('Estado de Destrui√ß√£o')
plt.ylabel('Frequ√™ncia')

# --- GR√ÅFICO 3: Localiza√ß√£o (Foco no Midwest) ---
plt.subplot(2, 2, 3)
# Ordena√ß√£o para mostrar os estados com mais ca√ßadas
ordem_estados = df_spn['Localizacao_Estado'].value_counts().index
sns.countplot(data=df_spn, x='Localizacao_Estado', order=ordem_estados, palette=['#96705B','#D1BEB0','#F7D4BC','#8D5B4C'])
plt.title('3. Geografia das Ca√ßadas (Foco no Midwest)', fontsize=14)
plt.xlabel('Estado dos EUA')
plt.ylabel('N√∫mero de Ca√ßadas')

# --- GR√ÅFICO 4: Besti√°rio (Tipos de Criatura) ---
plt.subplot(2, 2, 4)
ordem_criaturas = df_spn['Tipo_Criatura'].value_counts().index
sns.countplot(data=df_spn, x='Tipo_Criatura', order=ordem_criaturas, palette=['#4059AD','#07004D','#A13D63', '#7F0799'])
plt.title('4. Besti√°rio: Categorias de Inimigos', fontsize=14)
plt.xlabel('Tipo de Criatura')
plt.ylabel('Frequ√™ncia')

plt.tight_layout()
plt.savefig('analise_supernatural.png') # Guarda a imagem para o teu relat√≥rio
plt.show()
```

**Explica√ß√£o das Visualiza√ß√µes:**

*Impacto Emocional (Mortes):* Este gr√°fico de barras mostra que, √† medida que a s√©rie avan√ßa para o final da 5¬™ temporada, o n√∫mero de epis√≥dios com mortes importantes ($Sim$) supera os epis√≥dios "tranquilos" ($N√£o$). √â uma m√©trica de qu√£o dram√°tica a s√©rie se tornou.

*Estado do Impala:* Aqui usei a fun√ß√£o lambda para filtrar o estado "Destru√≠do". Visualmente, mostra que o carro √© quase um personagem: ele sobrevive a quase tudo, mas quando √© destru√≠do (como no final da Temporada 5), o impacto visual no gr√°fico √© imediato.

*Localiza√ß√£o (Midwest):* Todos os estados na lista (Kansas, South Dakota, Indiana, Illinois) pertencem √† regi√£o do Midwest americano. O gr√°fico mostra que o Kansas √© o "quartel-general" emocional dos irm√£os.

*Besti√°rio:* Esta visualiza√ß√£o separa os inimigos por hierarquia. Nota como come√ßa com um "Esp√≠rito" no Piloto e termina com "Anjos/L√∫cifer" no final, mostrando a escalada de poder (Power Creep) na s√©rie atrav√©s dos dados.

**üèÅ Conclus√£o e Insights**

Ap√≥s a an√°lise dos dados, conseguimos extrair conclus√µes estat√≠sticas que validam a narrativa da s√©rie:

*Escalada de Poder:* Atrav√©s da visualiza√ß√£o do "Besti√°rio", √© evidente a transi√ß√£o de monstros cl√°ssicos (fantasmas/lendas urbanas) para entidades c√≥smicas (anjos/dem√≥nios), o que justifica o aumento das notas no IMDb.

*Rela√ß√£o Nota vs. Trag√©dia:* Os dados mostram que os epis√≥dios com maior pontua√ß√£o (ex: Swan Song) coincidem com momentos de destrui√ß√£o do Impala ou morte de personagens importantes, provando que o impacto emocional √© um fator chave no sucesso da obra.

*Consist√™ncia Geogr√°fica:* A an√°lise confirmou que a maioria das "ca√ßadas" analisadas ocorre no Midwest americano, refor√ßando a identidade est√©tica da s√©rie.
